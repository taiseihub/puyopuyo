```python
e_files = ["/content/drive/MyDrive/E sports/PuyoJSON/S102.json", "/content/drive/MyDrive/E sports/PuyoJSON/S103.json", "/content/drive/MyDrive/E sports/PuyoJSON/S104.json", "/content/drive/MyDrive/E sports/PuyoJSON/S105.json", "/content/drive/MyDrive/E sports/PuyoJSON/S106.json", "/content/drive/MyDrive/E sports/PuyoJSON/S107.json", "/content/drive/MyDrive/E sports/PuyoJSON/S108.json", "/content/drive/MyDrive/E sports/PuyoJSON/S201.json", "/content/drive/MyDrive/E sports/PuyoJSON/S202.json", "/content/drive/MyDrive/E sports/PuyoJSON/S203.json", "/content/drive/MyDrive/E sports/PuyoJSON/S204.json", "/content/drive/MyDrive/E sports/PuyoJSON/S205.json", "/content/drive/MyDrive/E sports/PuyoJSON/S301.json", "/content/drive/MyDrive/E sports/PuyoJSON/S302.json", "/content/drive/MyDrive/E sports/PuyoJSON/S303.json", "/content/drive/MyDrive/E sports/PuyoJSON/S304.json", "/content/drive/MyDrive/E sports/PuyoJSON/S305.json", "/content/drive/MyDrive/E sports/PuyoJSON/S401.json", "/content/drive/MyDrive/E sports/PuyoJSON/S402.json", "/content/drive/MyDrive/E sports/PuyoJSON/S403.json", "/content/drive/MyDrive/E sports/PuyoJSON/S404.json"]
for i in e_files:
    with open(i) as f:
        j = json.load(f)
        print(j["compe"]["cid"])

        #連鎖最大値
        for players in range(2):
            x = 0
            rensa_list = []
            while True:
                try:
                    rensa_list.append(j["matches"][0]["games"][0]["players"][players]["rensas"][x]["count"])
                except IndexError:
                    break
                else:
                    x += 1
            
            if j["matches"][0]["winner"] == players:
                print(j["matches"][0]["match_info"]["players"][players]["name"] + "(winner)")
            if j["matches"][0]["winner"] != players:
                print(j["matches"][0]["match_info"]["players"][players]["name"])
            if len(rensa_list) != 0:
                print(f"最大連鎖:{max(rensa_list)}")
            if len(rensa_list) == 0:
                print("連鎖なし")
        
            #連鎖最頻値
            counter = collections.Counter(str(x))
            mode_v = counter.most_common()[0][-1]
            it = itertools.takewhile(
            lambda kv: kv[-1] == mode_v, counter.most_common()
            )
            print("最頻連鎖:", end="")
            print(*(k for k, v in it))
            print("")

            #描画
            if players == 0:
                rensa_t_list0 = []
                score_list0 = []
            if players == 1:
                rensa_t_list1 = []
                score_list1 = []           
            y = 0
            while True:
                try:
                    if players == 0:
                        rensa_t_list0.append(j["matches"][0]["games"][0]["players"][0]["triggers"][y]["t"])
                        score_list0.append(j["matches"][0]["games"][0]["players"][0]["triggers"][y]["score"])
                    else:
                        rensa_t_list1.append(j["matches"][0]["games"][0]["players"][1]["triggers"][y]["t"])
                        score_list1.append(j["matches"][0]["games"][0]["players"][1]["triggers"][y]["score"])
                except IndexError:
                    break
                else:
                    y += 1
        fig, ax = plt.subplots(facecolor="white")
        ax.plot(rensa_t_list0, score_list0, label=str(j["matches"][0]["match_info"]["players"][0]["name"]), color="red", marker="o")
        ax.plot(rensa_t_list1, score_list1, label=str(j["matches"][0]["match_info"]["players"][1]["name"]), color="blue", marker="v")
        ax.set_xlabel("発火時間")
        ax.set_ylabel("スコア", rotation="horizontal")
        ax.set_title(j["compe"]["cid"])
        ax.grid()
        ax.legend()

```
